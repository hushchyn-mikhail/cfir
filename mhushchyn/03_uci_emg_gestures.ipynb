```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

import pickle
```

# Read data

Data: https://archive.ics.uci.edu/ml/datasets/EMG+data+for+gestures

Paper: https://pdfs.semanticscholar.org/a680/d59544c868c26b25bfd5e4f1751cb6a4c4b7.pdf

```python
data = pd.read_csv("data/EMG_data_for_gestures-master/01/1_raw_data_13-12_22.03.16.txt", sep='\t')
data *= 10**6

data.head()
```

```python
X = data.values[:, 1:-1].astype(np.float32)
y = data.values[:, -1].astype(np.int)

#y = np.array([int(i) for i in y])
```

# Analysis

```python
for i_col in range(X.shape[1]):
    
    ax = X[:, i_col]
    lab = y #* 600 + ax.mean() - 300
    
    plt.figure(figsize=(12, 5))
    plt.plot(range(len(ax)), ax, label='channel_'+str(i_col+1), linewidth=2)
    #plt.plot(range(len(ax)), lab, label='Label', linewidth=1, color='0', linestyle='--')
    #plt.plot([0, len(ax)], [0, 0], color='0', linestyle='--', linewidth=2)
    
    plt.xlim(0, len(ax))
    plt.xticks(size=14)
    plt.xlabel('Measurement number', size=14)
    
    #plt.ylim(ax.mean()-200, ax.mean()+200)
    plt.yticks(size=14)
    plt.ylabel('Measurement value', size=14)
    
    #plt.title("Dataset "+aname, size=14, loc='left')
    plt.legend(loc='upper right', fontsize=14)
    
    plt.tight_layout()
    plt.savefig(fname="pics/03_uci_emg_gestures_"+'channel_'+str(i_col+1)+".pdf", format="pdf")
    plt.show()
```
