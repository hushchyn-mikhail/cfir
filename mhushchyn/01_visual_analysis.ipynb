```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

import pickle
```

# Read data

```python
with open("rest_state_probes_real.pkl", 'rb') as pickleFile:
    data = pickle.load(pickleFile)
    
data.head()
```

```python
data.shape
```

# Analysis

```python
# Number of unique datasets
print("Unique datasets: ")
np.unique(data['dataset'].values)
```

```python
for aname in np.unique(data['dataset'].values):
    adataset = data[data['dataset'].values == aname]
    eeg = adataset['eeg'].values
    
    plt.figure(figsize=(12, 5))
    plt.plot(eeg, label='EEG')
    plt.plot([0, len(eeg)], [0, 0], color='0', linestyle='--', linewidth=2)
    
    plt.xlim(0, len(eeg))
    plt.xticks(size=14)
    plt.xlabel('Measurement number', size=14)
    
    plt.ylim(-0.00004, 0.00004)
    plt.yticks(size=14)
    plt.ylabel('Measurement value', size=14)
    
    plt.title("Dataset "+aname, size=14, loc='left')
    plt.legend(loc='upper right', fontsize=14)
    
    plt.tight_layout()
    plt.savefig(fname="pics/egg_full_"+aname+".pdf", format="pdf")
    plt.show()
```
